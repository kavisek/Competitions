# <center> Descriptive Analysis

### Import Prelminaries

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

train_df = pd.read_csv('https://www.dropbox.com/s/nqftja8aa01z4mn/cleaned_train_df.csv?dl=1', index_col='Id')
test_df = pd.read_csv('https://www.dropbox.com/s/gw7fi8o71kczq6q/cleaned_test_df.csv?dl=1', index_col='Id')

print('Training Dataframe:', train_df.shape)
print('Test Dataframe:', test_df.shape)
```

```python
# View the dataframe
train_df.head(5)
```

### Quantative Values Histogram Values

```python
for col in train_df.select_dtypes(['int64','float64']):
  plt.figure()
  train_df[col].hist(color='black',grid=False, edgecolor='w' )
  plt.title(str(train_df[col].name + ' Histogram'))
  plt.xlabel(str(train_df[col].name ))
  plt.ylabel('Frequency')
  
  plt.axvline(train_df[col].mean(), color='r', 
              linestyle='dashed', linewidth=2)
```

### Quantative Correlation Matrix

```python
def df_palette(val):
  color = 'darkred' if val < 0 else 'green'
  if val == 0 or val == 1: color = 'black' 
  return 'color: %s'%color
```

```python
corr_matrix = train_df.select_dtypes(['int64','float64']).corr()
corr_matrix.style.applymap(df_palette)
```

### Qualitative Histogram Counts

```python
for col in train_df.select_dtypes('object'):
  df = pd.DataFrame(train_df[col].value_counts()).plot.barh(
  color='darkred',edgecolor='w')
  df.columns = ['Count']
  plt.title(str(train_df[col].name) + "Count")
  plt.xlabel(str(train_df[col].name))
  plt.ylabel('Frequency')
```
